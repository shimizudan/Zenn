```julia
using StatsBase
function saikoro5()
    a = rand(1:6)
    if a != 6
        return a
    else saikoro5()
    end
end

[saikoro5() for _=1:5*10^6] |> countmap
```

```julia
using StatsBase
function saikoro7()
    a,b = rand(1:6),rand(1:6)
    if a != b
        return a
    elseif a == b != 6
        return 7
    else saikoro7()
    end
end

[saikoro7() for _=1:7*10^6] |> countmap
```
